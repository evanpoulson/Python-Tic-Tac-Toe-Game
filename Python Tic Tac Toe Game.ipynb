```python
from IPython.display import clear_output
from random import randint
```

```python
#function that takes a list as an input, and displays the list as a tic tac toe board
#game function
def display_board(board_list):
    
    layer_size = 3
    board_display = []
    
    #iterate through the list and split it into layers of three 
    for i in range(0, len(board_list), layer_size):
        board_display.append(board_list[i:i + layer_size])
    
    #print each layer of the board
    for layer in board_display:
        print(layer)
```

```python
#function that assigns a users choice to a position on the board
#game function
def place_marker(board, marker, position):
    
    numpad_list = ['7','8','9','4','5','6','1','2','3']
    
    #loop that finds the index of the position in the numpad_list
    choice_index = 0
    for num in numpad_list:
        if num == position:
            break
        else:
            choice_index += 1
    
    board[choice_index] = marker
    return board
```

```python
#function that checks if anyone has won the game after every move
#game function
def win_check(board,mark):
    
     return ((board[6] == mark and board[7] == mark and board[8] == mark) or # across the top
    (board[3] == mark and board[4] == mark and board[5] == mark) or # across the middle
    (board[0] == mark and board[1] == mark and board[2] == mark) or # across the bottom
    (board[6] == mark and board[3] == mark and board[0] == mark) or # down the middle
    (board[7] == mark and board[4] == mark and board[1] == mark) or # down the middle
    (board[8] == mark and board[5] == mark and board[2] == mark) or # down the right side
    (board[6] == mark and board[4] == mark and board[2] == mark) or # diagonal
    (board[8] == mark and board[4] == mark and board[0] == mark)) # diagonal
```

```python
#function that determines wether or not a space on the board is free, or filled
#game function
def space_check(board,position):
    
    numpad_list = ['7','8','9','4','5','6','1','2','3']
    
    #loop that finds the index of the position in the numpad_list
    choice_index = 0
    for num in numpad_list:
        if int(num) == int(position):
            break
        else:
            choice_index += 1
    
    if board[choice_index] == ' ':
        return True
    else:
        return False
```

```python
#function that checks if the board is full, and therefore a tie has occured
#game function
def board_full(board):
    
    filled_spots = 0
    
    #iterate through the board and count how many filled spaces there are
    for position in board:
        if position != ' ':
            filled_spots += 1
        else:
            pass
    
    if filled_spots == 9:
        return True
    else:
        return False
```

```python
#write a function that asks the player for their position choice, and checks if it is viable
#game function
def player_choice(board):
    
    numpad_list = ['7','8','9','4','5','6','1','2','3']
    position = 0
    
    while position not in numpad_list or not space_check(board,position):
        position = input("Choose a position (1-9): ")
    
    return str(position)
```

```python
#function that assigns the players marker as either X or O
#setup function
def player_marker():
    
    valid_inputs = ['X','O']
    player_1_marker = ' '
    
    #input to decide player1s marker
    while player_1_marker not in valid_inputs:
        player_1_marker = input("Player 1, would you like to be X or O? ").upper()
        
        if player_1_marker not in valid_inputs:
            clear_output(True)
            print('Sorry, invalid input! Please try again!')
            
        
    if player_1_marker == 'X':
        return 'X','O'
    else:
        return 'O','X'
```

```python
#function that determins who moves first in the game
#setup function
def first_move():
    
    num = randint(0,1)
    
    if num == 0:
        return 'Player1'
    else:
        return 'Player2'
```

```python
#function that asks user if they want to replay the game
#setup function
def replay():
    
    valid_choices = ['Y','N']
    choice = 0
    
    while choice not in valid_choices:
        choice = input('Would you like to play again? (Y/N): ').upper()
    
    if choice == 'Y':
        return True
    else:
        return False
```

```python
#function that starts the game
#setup function
def game_on():
    
    viable_choices = ['Y','N']
    choice = 0
    
    while choice not in viable_choices:
        choice = input('Are you ready to begin playing? (Y/N): ').upper()
    
    if choice == 'Y':
        return True
    else:
        return False
```

```python
#all the functions together to create a working tic tac toe game in python! Hooray!
print('Welcome to Tic Tac Toe in Python!')
while True:
    board = [' '] * 9
    player1_marker,player2_marker = player_marker()
    
    turn = first_move()
    if turn == 'Player1':
        print('Player1 will go first')
    else:
        print('Player2 will go first')
    
    game_choice = game_on()
    
    while game_choice:
        
        #player1 turn
        if turn == 'Player1':
           
            display_board(board)
            position = player_choice(board)
            place_marker(board, player1_marker, position)

            if win_check(board, player1_marker):
                display_board(board)
                print('Congratulations! You have won the game!')
                game_on = False
                break
            else:
                if board_full(board):
                    display_board(board)
                    print('The game is a draw!')
                    break
                else:
                    turn = 'Player 2'
        else:
            
            display_board(board)
            position = player_choice(board)
            place_marker(board, player2_marker, position)

            if win_check(board, player2_marker):
                display_board(board)
                print('Player 2 has won!')
                game_on = False
                break
            else:
                if board_full(board):
                    display_board(board)
                    print('The game is a draw!')
                    break
                else:
                    turn = 'Player1'

    if not replay():
        break
```
